```python
%%writefile word_count_mr.py
from mrjob.job import MRJob
import re
#I was born on 23 july 2002
WORD_RE = re.compile(r'\b\w+\b')

class WordCount(MRJob):

    def mapper(self, _, line):
        for word in WORD_RE.findall(line.lower()):
            yield word, 1

    def reducer(self, word, counts):
        yield word, sum(counts)

if __name__ == "__main__":
    WordCount.run()
```

```python
%%writefile non_english_mr.py
from mrjob.job import MRJob
from spellchecker import SpellChecker
import re

WORD_RE = re.compile(r'\b\w+\b')

class NonEnglishWordCount(MRJob):

    def mapper_init(self):
        self.spell_checker = SpellChecker()

    def mapper(self, _, line):
        for word in WORD_RE.findall(line.lower()):
            if word not in self.spell_checker:
                yield word, 1

    def reducer(self, word, counts):
        yield word, sum(counts)

if __name__ == "__main__":
    NonEnglishWordCount.run()
```
